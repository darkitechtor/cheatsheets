```python
"""
import libraries: codecs - for reading html file,
and Selector - for getting nesessary parts of html file
"""
import pandas as pd
import codecs
from scrapy import Selector
```

```python
# read html downloaded file
page = codecs.open('/content/drive/MyDrive/Colab Data/Транзакции.html', 'r',
    'utf-8')
html = page.read()
```

```python
# construct a Selector
sel = Selector(text = html)
```

```python
# get comments
names = sel.css('div.comment::text').extract()
# get signs
signs = sel.css('div.balance > span.debt > span > span.sign::text').extract()
# get amounts
sums = sel.css('div.balance > span.debt > span > span.sum::text').extract()
```

```python
# create df
df = pd.DataFrame(names) 
df[1] = signs
df[2] = sums
# rename columns
df.columns = ['name','sign','amount']
# check
df
```

```python
#
df_plus = df[df['sign']=='+']
df_minus = df[df['sign']=='−']
#
df_final = df_minus.merge(df_plus, on='name', how='outer')
df_final
```

```python
#save to csv
df_final.to_csv()
```

```
"""
import libraries:
codecs - for reading html file
Selector - for getting nesessary parts of html file
"""
import pandas as pd
from Scrapy import Selector
import requests
```

```
# read html downloaded file
response = requests.get("https://stock.toyota.ru/new/toyota?cities=krasnodar&reversed=false")
if response.status_code != 200:
	print("Error fetching page")
	exit()
else:
	content = response.content
print(content)
```

```
# construct a Selector
sel = Selector(text=content)
sel
```

```
# get
names = sel.css("tr-newcars-models > div.tr-page-body > aside > tr-scroll > div.tr-scroll > tr-new-car-filter > div > form > div > tr-filter-accordion:nth-child(1) > button > span.tr-h5.tr-filter-head-caption > span").extract()
names
```
